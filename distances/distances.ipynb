# Distribution of distances between random points

We empirically study the distribution of distances between random points in the plane.

```python
import numpy as np
import matplotlib.pyplot as plt
import sys
from distances import min_distances, min_indexes, pairwise_distances_euclidean, pairwise_distances_manhattan

%matplotlib inline
```

## Histogram of distances between random points
We draw points at random uniformly in the square $[0,1]^2$, then compute the minimum pairwise distances for each point and finally plot the distribution. We use Euclidian and Manhattan distance.

```python
size = 10000
points = np.random.random_sample(size=(2, size))
```

```python
min_dists_euclidean = min_distances(points, distance='euclidean')
min_dists_manhattan = min_distances(points, distance='manhattan')
```

```python
plt.hist(min_dists_euclidean, bins='auto');
```

```python
plt.hist(min_dists_manhattan, bins='auto');
```

## Graphs of minimum distances
We draw points at random uniformly in the square $[0,1]^2$ and compute the minimum pairwise distances for each point. Then, we connect each point to its nearest neighbor. 

```python
size = 150
points = np.random.random_sample(size=(2, size))
```

```python
pdists_eucl = pairwise_distances_euclidean(points)
pdists_eucl[pdists_eucl == 0] = np.inf
pdists_manh = pairwise_distances_manhattan(points)
pdists_manh[pdists_manh == 0] = np.inf
```

```python
min_ind_eucl = min_indexes(pdists_eucl)
min_ind_manh = min_indexes(pdists_manh)
```

```python
plt.figure(figsize=(10, 10))
xs, ys = points
plt.scatter(xs, ys)
for i in range(size):
    idx = (min_ind_eucl['row_idx'][i], min_ind_eucl['col_idx'][i])
    plt.plot(xs[[idx]], ys[[idx]], c='r')
plt.title('nearest 1-neighbor (Euclidian distance)');
```

```python
plt.figure(figsize=(10, 10))
plt.scatter(xs, ys)
for i in range(size):
    idx = (min_ind_manh['row_idx'][i], min_ind_manh['col_idx'][i])
    plt.plot(xs[[idx]], ys[[idx]], c='r')
    plt.title('nearest 1-neighbor (Manhattan distance)');
```
